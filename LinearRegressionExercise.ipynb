```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
dt = pd.read_csv('/content/canada_per_capita_income.csv')
dt.head()
```

```python
new_dt = dt.drop('per capita income (US$)',axis='columns')
new_dt
```

```python
model=linear_model.LinearRegression()
model.fit(new_dt,dt['per capita income (US$)'])
```

```python
model.predict([[2020]])
```
